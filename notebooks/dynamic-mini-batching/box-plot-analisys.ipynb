```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando os dados do CSV para um DataFrame
url = '/Users/reginaldo.junior/Documents/Ufscar/tecnhiques-reduce-consumption-energy/results/mb_socket_dynamic.csv'  # Substitua pelo caminho do seu arquivo CSV
df = pd.read_csv(url)
df = df[df.dataset == 'airlines']
display(df)

# Criando o grÃ¡fico de caixa com Seaborn
plt.figure(figsize=(8, 10))

# Box plot para a coluna "delay"
plt.subplot(3, 1, 1)
sns.boxplot(x='algorithm', y='JPI', data=df)
plt.title('JPI')

# Box plot para a coluna "delay"
plt.subplot(3, 1, 2)
sns.boxplot(x='algorithm', y='delay_s', data=df)
plt.title('Delay')

# Box plot para a coluna "IPS"
plt.subplot(3, 1, 3)
sns.boxplot(x='algorithm', y='IPS', data=df)
plt.title('IPS')

# Ajustando o layout
plt.tight_layout()

# Exibindo o grÃ¡fico
plt.show()
```
